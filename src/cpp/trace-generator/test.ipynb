```python
fname = "share-gpt-status.tsv"
# fname = "alpaca-status.tsv"
dataset = import_tsv(fname)
batched_request = BatchedRequest(dataset=dataset)

batch_size = 256
refill_duration = 20

for i in range(30000):
    batched_request.cycle(refill_duration)
    batched_request.fill(batch_size)
    num_cycles, sequences = batched_request.snapshot()

    # print(f"cycle count {num_cycles} : {sequences}")
    print(f"cycle count {num_cycles}")
    print(f"average sequence length: {sum(sequences)/len(sequences)}")
    if i > 10000:
        break
```
